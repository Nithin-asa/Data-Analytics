**TASK 03** - PLOTTING THE DISTRIBUTION OF VARIOUS DISCRETE FEATURES ON SEASON/HOLIDAY/WEEKEND/WEATHERCODE

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ... (your previous code) ...

# Define discrete_features - Replace with your actual discrete feature names
discrete_features = ['cnt', 't1', 't2', 'hum', 'wind_speed', 'weather_code', 'is_holiday', 'is_weekend', 'season']

def plot_feature_distribution(features, data):
    # Calculate the number of rows and columns for subplots
    num_features = len(features)
    num_cols = 2  # You can adjust this
    num_rows = (num_features + num_cols - 1) // num_cols  # Calculate rows needed

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 10))  # Create subplots with calculated dimensions
    axes = axes.flatten()  # Flatten for easier iteration

    for i, feature in enumerate(features):
        if feature in data.columns:
            sns.histplot(data[feature], ax=axes[i])
            axes[i].set_title(f'Distribution of {feature}')
            axes[i].set_xlabel(feature)
            axes[i].set_ylabel('Count')
        else:
            print(f"Column '{feature}' not found in DataFrame. Skipping...")

    # Hide any unused subplots (if num_features is not a multiple of num_cols * num_rows)
    for i in range(num_features, len(axes)):
        axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()

plot_feature_distribution(discrete_features, data)
```

**TASK 04** - LOOKING AT THE DATA TYPES OF EACH VARIABLE, TRANSFORM TIMESTAMP IN TYPE & SET IT AS INDEX

```python
# prompt:  LOOKING AT THE DATA TYPES OF EACH VARIABLE, TRANSFORM TIMESTAMP IN TYPE & SET IT AS INDEX geneate code

# Check if the column name is 'timestamp' or something else (e.g., 'date')
timestamp_column_name = 'timestamp'  # Replace with the actual column name if different

# Convert the timestamp column to datetime objects
if timestamp_column_name in data.columns:
    data[timestamp_column_name] = pd.to_datetime(data[timestamp_column_name])

    # Set the timestamp column as the index
    data = data.set_index(timestamp_column_name)
else:
    print(f"Column '{timestamp_column_name}' not found in DataFrame. Please check the column name.")
```

**TASK 05** - MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC

```python
# prompt: MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC

# **TASK 05** - MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC

# Assuming your timestamp column is now the index, you can directly access the date/time components:
data['DayOfWeek'] = data.index.dayofweek
data['DayOfMonth'] = data.index.day
data['Hour'] = data.index.hour
data['Month'] = data.index.month
data['Year'] = data.index.year

# Season is already in your dataset, but you can create a more descriptive mapping if needed.
# Example mapping (replace with your actual season definitions):
#season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Autumn', 4: 'Winter'}
#data['SeasonName'] = data['season'].map(season_mapping)

# Example for creating a "Weekend" feature
data['IsWeekend'] = (data['DayOfWeek'] >= 5).astype(int)

# Example for creating a feature for the time of day (morning, afternoon, evening, night)
# Customize the thresholds as per your requirement
def get_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'

data['TimeOfDay'] = data['Hour'].apply(get_time_of_day)


# Display the updated DataFrame with the new features
print(data.head())
```

**TASK 06** - VISUALIZING THE CORRELATION WITH A HEATMAP

```python
# prompt: VISUALIZING THE CORRELATION WITH A HEATMAP

# ... (your previous code) ...

# **TASK 06** - VISUALIZING THE CORRELATION WITH A HEATMAP
plt.figure(figsize=(12, 10))

# Select only numerical features for correlation calculation
numerical_data = data.select_dtypes(include=['number'])

sns.heatmap(numerical_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()
```

**TASK 07** - VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLO

```python
# prompt: VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLO

# ... (your previous code) ...

# **TASK 07** - VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLOT

# Assuming 'cnt' is your target variable
target_variable = 'cnt'

# Calculate correlations with the target variable
# Select only numerical features for correlation calculation
numerical_data = data.select_dtypes(include=['number'])  # Filter out non-numerical columns

correlations = numerical_data.corr()[target_variable].drop(target_variable)  # Drop the target variable itself

# Create a barplot of the correlations
plt.figure(figsize=(10, 6))
correlations.plot(kind='bar')
plt.title(f'Correlation of Features with {target_variable}')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**TASK 10** - PLOTTING BIKE SHARES BY HOURS ON (HOLIDAYS, WEEKEND, SEASON)

```python
# prompt: PLOTTING BIKE SHARES BY HOURS ON (HOLIDAYS, WEEKEND, SEASON)

# **TASK 10** - PLOTTING BIKE SHARES BY HOURS ON (HOLIDAYS, WEEKEND, SEASON)

# Plotting bike shares by hour for holidays
plt.figure(figsize=(10, 6))
sns.lineplot(x='Hour', y='cnt', data=data[data['is_holiday'] == 1], label='Holiday')
plt.title('Bike Shares by Hour on Holidays')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Bike Shares')
plt.show()


# Plotting bike shares by hour for weekends
plt.figure(figsize=(10, 6))
sns.lineplot(x='Hour', y='cnt', data=data[data['is_weekend'] == 1], label='Weekend')
plt.title('Bike Shares by Hour on Weekends')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Bike Shares')
plt.show()

# Plotting bike shares by hour for each season
plt.figure(figsize=(10, 6))
for season in data['season'].unique():
    sns.lineplot(x='Hour', y='cnt', data=data[data['season'] == season], label=f'Season {season}')
plt.title('Bike Shares by Hour for Each Season')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Bike Shares')
plt.legend()
plt.show()
```

**TASK 14** - VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

```python
# prompt: VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

# **TASK 14** - VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

# Calculate the total bike shares for weekdays and weekends
weekday_shares = data[data['IsWeekend'] == 0]['cnt'].sum()
weekend_shares = data[data['IsWeekend'] == 1]['cnt'].sum()

# Create a pie chart
labels = ['Weekday', 'Weekend']
sizes = [weekday_shares, weekend_shares]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Bike Shares by Weekday/Weekend (Pie Chart)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Create a bar plot
plt.figure(figsize=(6, 4))
plt.bar(labels, sizes, color=['skyblue', 'lightcoral'])
plt.title('Distribution of Bike Shares by Weekday/Weekend (Bar Plot)')
plt.xlabel('Day Type')
plt.ylabel('Total Bike Shares')
plt.show()
```

**TASK 15** - PLOTTING THE DISTRIBUTION OF WEATHER CODE BY SEASON

```python
# prompt: PLOTTING THE DISTRIBUTION OF WEATHER CODE BY SEASON

# **TASK 15** - PLOTTING THE DISTRIBUTION OF WEATHER CODE BY SEASON

plt.figure(figsize=(12, 6))
sns.countplot(x='weather_code', hue='season', data=data)
plt.title('Distribution of Weather Code by Season')
plt.xlabel('Weather Code')
plt.ylabel('Count')
plt.show()
```

**TASK 09** - PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)

```python
# prompt: generate a code for PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)

# Plotting Bike Shares by Month
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='cnt', data=data)
plt.title('Bike Shares by Month (Line Plot)')
plt.xlabel('Month')
plt.ylabel('Number of Bike Shares')
plt.show()

plt.figure(figsize=(10, 6))
sns.pointplot(x='Month', y='cnt', data=data)
plt.title('Bike Shares by Month (Point Plot)')
plt.xlabel('Month')
plt.ylabel('Number of Bike Shares')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='cnt', data=data, ci=None) # ci=None removes confidence intervals
plt.title('Bike Shares by Month (Bar Plot)')
plt.xlabel('Month')
plt.ylabel('Number of Bike Shares')
plt.show()


# Plotting Bike Shares by Year and Month (using a combined approach)
plt.figure(figsize=(12, 6))
for year in data['Year'].unique():
    sns.lineplot(x='Month', y='cnt', data=data[data['Year'] == year], label=str(year))
plt.title('Bike Shares by Month and Year')
plt.xlabel('Month')
plt.ylabel('Number of Bike Shares')
plt.legend(title='Year')
plt.show()
```
