```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Load dataset
df = pd.read_csv("Games.csv")
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

# Initialize Dash app
app = dash.Dash(__name__)
app.layout = html.Div([
    dcc.Input(id='console-input', type='text', placeholder='Enter Console (e.g., PC, PS4)') ,
    dcc.Graph(id='score-distribution'),
    html.Div(id='error-message', style={'color': 'red'})
])

@app.callback(
    [Output('score-distribution', 'figure'), Output('error-message', 'children')],
    [Input('console-input', 'value')]
)
def update_graph(console):
    if not console:
        return px.histogram(), "Please enter a console."

    console = console.lower()
    filtered_data = df[df['console'].str.lower().str.contains(console, na=False)]

    if filtered_data.empty:
        return px.histogram(), f"No data found for {console.upper()}."

    fig = px.histogram(filtered_data, x='score', nbins=10, title=f"Score Distribution for {console.upper()}",
                        labels={'score': 'Review Score', 'count': 'Number of Games'})

    return fig, ""

if __name__ == '__main__':
    app.run_server(debug=True)
```
