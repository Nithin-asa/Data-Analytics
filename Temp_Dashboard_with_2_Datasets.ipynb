```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Load dataset
df = pd.read_csv("GlobalWeatherRepository.csv")
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

# Initialize Dash app
app = dash.Dash(__name__)
app.layout = html.Div([
    dcc.Input(id='country-input', type='text', placeholder='Enter Country Name'),
    dcc.Graph(id='temperature-distribution'),
    html.Div(id='error-message', style={'color': 'red'})
])

@app.callback(
    [Output('temperature-distribution', 'figure'), Output('error-message', 'children')],
    [Input('country-input', 'value')]
)
def update_graph(country):
    if not country:
        return px.histogram(), "Please enter a country name."

    country = country.lower()
    filtered_data = df[df['country'].str.lower().str.contains(country, na=False)]

    if filtered_data.empty:
        return px.histogram(), f"No data found for {country.upper()}."

    fig = px.histogram(filtered_data, x='temperature_celsius', nbins=20,
                        title=f"Temperature Distribution in {country.upper()}",
                        labels={'temperature_celsius': 'Temperature (°C)', 'count': 'Number of Records'})

    return fig, ""

if __name__ == '__main__':
    app.run_server(debug=True)
```

The above output shows Global Weather Repository

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Load the city_temperature.csv dataset
df = pd.read_csv('city_temperature.csv')

# Data Cleaning and Preparation
# Handle missing values and convert temperature to Celsius if needed
df = df.dropna()
df['AvgTemperature'] = df['AvgTemperature'].astype(float)

# Convert Fahrenheit to Celsius if necessary
# (Assuming AvgTemperature is in Fahrenheit in the dataset)
df['AvgTemperatureCelsius'] = (df['AvgTemperature'] - 32) * 5/9

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Average Temperature Distribution Across Cities"),
    dcc.Dropdown(
        id='city-dropdown',
        options=[{'label': city, 'value': city} for city in df['City'].unique()],
        value=df['City'].unique()[0],  # Set default value
        multi=False
    ),
    dcc.Graph(id='temperature-pie-chart'),
])

@app.callback(
    Output('temperature-pie-chart', 'figure'),
    Input('city-dropdown', 'value')
)
def update_pie_chart(selected_city):
    filtered_df = df[df['City'] == selected_city]

    fig = px.pie(
        filtered_df,
        values='AvgTemperatureCelsius',
        names='Region',  # Assuming you want to group by Region within the city
        title=f"Average Temperature Distribution in {selected_city}"
    )

    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
# prompt: import dash
# from dash import dcc, html
# from dash.dependencies import Input, Output
# import plotly.express as px
# import pandas as pd
# # Load the city_temperature.csv dataset
# df = pd.read_csv('city_temperature.csv')
# # Data Cleaning and Preparation
# # Handle missing values and convert temperature to Celsius if needed
# df = df.dropna()
# df['AvgTemperature'] = df['AvgTemperature'].astype(float)
# # Convert Fahrenheit to Celsius if necessary
# # (Assuming AvgTemperature is in Fahrenheit in the dataset)
# df['AvgTemperatureCelsius'] = (df['AvgTemperature'] - 32) * 5/9
# app = dash.Dash(__name__)
# app.layout = html.Div([
#     html.H1("Average Temperature Distribution Across Cities"),
#     dcc.Dropdown(
#         id='city-dropdown',
#         options=[{'label': city, 'value': city} for city in df['City'].unique()],
#         value=df['City'].unique()[0],  # Set default value
#         multi=False
#     ),
#     dcc.Graph(id='temperature-pie-chart'),
# ])
# @app.callback(
#     Output('temperature-pie-chart', 'figure'),
#     Input('city-dropdown', 'value')
# )
# def update_pie_chart(selected_city):
#     filtered_df = df[df['City'] == selected_city]
#     fig = px.pie(
#         filtered_df,
#         values='AvgTemperatureCelsius',
#         names='Region',  # Assuming you want to group by Region within the city
#         title=f"Average Temperature Distribution in {selected_city}"
#     )
#     return fig
# if __name__ == '__main__':
#     app.run_server(debug=True)      i need scatter plot

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Load the city_temperature.csv dataset
df = pd.read_csv('city_temperature.csv')

# Data Cleaning and Preparation
# Handle missing values and convert temperature to Celsius if needed
df = df.dropna()
df['AvgTemperature'] = df['AvgTemperature'].astype(float)

# Convert Fahrenheit to Celsius if necessary
# (Assuming AvgTemperature is in Fahrenheit in the dataset)
df['AvgTemperatureCelsius'] = (df['AvgTemperature'] - 32) * 5/9

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Average Temperature Scatter Plot Across Cities"),
    dcc.Dropdown(
        id='city-dropdown',
        options=[{'label': city, 'value': city} for city in df['City'].unique()],
        value=df['City'].unique()[0],  # Set default value
        multi=False
    ),
    dcc.Graph(id='temperature-scatter-plot'),
])

@app.callback(
    Output('temperature-scatter-plot', 'figure'),
    Input('city-dropdown', 'value')
)
def update_scatter_plot(selected_city):
    filtered_df = df[df['City'] == selected_city]

    fig = px.scatter(
        filtered_df,
        x='Month',
        y='AvgTemperatureCelsius',
        color='Region',
        title=f"Average Temperature Scatter Plot in {selected_city}",
        labels={'AvgTemperatureCelsius': 'Average Temperature (°C)'}
    )

    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
